```python
class Node:
  def __init__(self,key):
    self.data = key
    self.left = None
    self.right = None

def height_tree(A):
  if (A == None):
    return 0
  else:
    # return max(height_tree(A.left), height_tree(A.right))+1
    return min(height_tree(A.left), height_tree(A.right))+1

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.left.left.left = Node(7)
root.right.right = Node(6)

print(height_tree(root))


```

```python
import operator
A = ["2","1","+","3","*"]
stack = []

ops = {
    '+': operator.add,
    '-': operator.sub,
    '*': operator.mul,
    '/': operator.truediv
}


op = ["+","-","*","/"]
for i in A:
  if i in op:
    op2 = int(stack.pop())
    op1 = int(stack.pop())
    i = i
    op3 = ops[i](op2,op1)
    stack.append(op3)

  else:
    stack.append(i)


print(stack)
```
