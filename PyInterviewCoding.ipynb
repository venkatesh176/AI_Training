```python
class Node:
  def __init__(self,key):
    self.data = key
    self.left = None
    self.right = None

def height_tree(A):
  if (A == None):
    return 0
  else:
    # return max(height_tree(A.left), height_tree(A.right))+1
    return min(height_tree(A.left), height_tree(A.right))+1

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.left.left.left = Node(7)
root.right.right = Node(6)

print(height_tree(root))


```

```python
def fib(n):
  count = 1
  first = 0
  second = 1

  while count<n-1:
    sum = first
    first =second
    second = first+ sum

    count= count+1
    yield sum


f = fib(10)

print(f.__next__())
print(f.__next__())
print(f.__next__())
print(f.__next__())
print(f.__next__())
print(f.__next__())
print(f.__next__())
print(f.__next__())



```
